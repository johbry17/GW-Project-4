```python
# import dependencies
import pandas as pd
from pathlib import Path
from sqlalchemy import create_engine
```

```python
# read in csv
Data_df = pd.read_csv(Path("ames_housing_data.csv"))

Data_df.head(20)
```

```python
# Create an SQLite database engine named "mydatabase.db"
engine = create_engine("sqlite:///database.db")

# Insert data into the SQLite database
Data_df.to_sql("houses", con=engine, index=False, if_exists="replace")

# Query the database to verify the insertion
query = "SELECT * FROM houses LIMIT 5"
result_df = pd.read_sql(query, engine)
result_df.head(20)
```
